```
# calculates normalized histogram
def imhist(img):
	m, n = img.shape
	h = [0.0] * 256
	for i in range(m):
		for j in range(n):
			h[img[i, j]] += 1
	return np.array(h) / (m * n)

# CDF
def cumsum(h):
	return [sum(h[:i+1]) for i in range(len(h))]

# main
def autoHist(img):
  # Contrast Stretching
  L = np.amin(img)
  H = np.amax(img)

  if(L != H):
    img = (255 * (img - L)) / (H - L)
  img = np.int_(img)
  return img
  # Histogram Equalisation
  h = imhist(img)
  cdf = np.array(cumsum(h))
  sk = np.uint8(255 * cdf)
  s1, s2 = img.shape
  Y = np.zeros_like(img)

  for i in range(0, s1):
    for j in range(0, s2):
      Y[i, j] = np.uint8(sk[img[i, j]])
  H = imhist(Y)

  # Gamma correction
  gamma = 0.5
  invGamma = 1.0 / gamma
  table = np.array([((i / 255.0) ** invGamma) * 255
    for i in np.arange(0, 256)]).astype("uint8")

  return cv2.LUT(Y, table)
```

```
plt.imshow(autoHist(image))
```
