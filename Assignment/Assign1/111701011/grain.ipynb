```
# input image
image = cv2.imread('grain.png')
#image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#(thresh, bwImage) = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY)
plt.imshow(image)
```

```
# applying gaussian filter to extract the smooth background
blur = image
for i in range(1000):
  blur = cv2.GaussianBlur(blur,(7,7),1)
plt.imshow(blur)
```

```
# applying division operation on each pixel to balance out the background
n, m, k = image.shape
for i in range(n):
  for j in range(m):
    for d in range(k):
      image[i][j][d] = image[i][j][d] / (blur[i][j][d] + 0.001) * 255.0
plt.imshow(image)
```
