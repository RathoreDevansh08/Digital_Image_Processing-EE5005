```
image = cv2.imread('white_rectangle.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```
#kernel = nm.ones((3,3), nm.float32) / float(9)
kernel = np.ndarray((3,3), np.float32)
kernel[0,0] = 1
kernel[0,1] = 0
kernel[0,2] = 0
kernel[1,0] = 0
kernel[1,1] = 0
kernel[1,2] = 0
kernel[2,0] = 0
kernel[2,1] = 0
kernel[2,2] = -1
dst = cv2.filter2D(image, -1, kernel)
plt.imshow(dst)
```
