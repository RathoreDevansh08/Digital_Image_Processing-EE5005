```
image = cv2.imread('W3-300x300.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```
print (nm.shape(image))
```

```
plt.imshow(image)
plt.show()
```

```
# Transformation:
x, y, z = nm.shape(image)
new_image = nm.ndarray(nm.shape(image))
for clr in range(z):
  for i in range(x):
    for j in range(y):
      new_image[i, j, clr] = ((-2.0) * image[i, j, clr])
      if (j != 0 ):
        new_image[i, j, clr] = new_image[i, j, clr] + image[i, (j-1), clr]
      if (j != (y-1)):
        new_image[i, j, clr] = new_image[i, j, clr] + image[i, (j+1), clr]
```

```
print(nm.shape(new_image))
```

```
plt.subplot(1, 2, 1)
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.imshow(new_image)

plt.show()
```

```
# Histogram plot

plt.subplot(1, 2, 1)
plt.hist(image.ravel(),256,[0,256])

plt.subplot(1, 2, 2)
plt.hist(new_image.ravel(),256,[0,256])

plt.show()
```
